```python
def truncate_texts(texts):
    """
    对输入的文本列表进行处理，若文本长度超过 256，则取最前的 200 个以及最后的 56 个字符组成新文本。

    :param texts: 文本列表，每个元素为一个文本字符串
    :return: 处理后的文本列表
    """
    return [text[:200] + text[-56:] if len(text) > 256 else text for text in texts]

df['文本'] = truncate_texts(df['文本'])
```

```python
import pandas as pd
from collections import Counter

def resolve_label_conflicts(df, text_col='文本', label_col='类别'):
    """
    解决重复文本的标注冲突：
    1. 优先选择该文本出现最频繁的标签（众数）
    2. 若频次相同，选择全局出现频率更高的标签
    3. 若全局频率相同，保留原始顺序的首个标签
    
    参数：
        df: 包含文本和标签的DataFrame
        text_col: 文本列名（默认'文本'）
        label_col: 标签列名（默认'类别'）
        
    返回：
        去重后的DataFrame（每个文本仅保留一行）
    """
    # 计算全局标签频率（备用）
    global_freq = df[label_col].value_counts().to_dict()
    
    # 定义冲突解决函数
    def resolve_group(group):
        if len(group) == 1:
            return group.iloc[0]  # 无冲突直接返回
            
        # 统计组内标签频次
        label_counts = group[label_col].value_counts()
        max_count = label_counts.max()
        top_labels = label_counts[label_counts == max_count].index.tolist()
        
        # 规则1：存在唯一高频标签
        if len(top_labels) == 1:
            return group[group[label_col] == top_labels[0]].iloc[0]
            
        # 规则2：按全局频率排序
        top_labels_sorted = sorted(
            top_labels, 
            key=lambda x: (-global_freq.get(x, 0), group.index.get_loc(group[group[label_col]==x].index[0]))
        )
        selected_label = top_labels_sorted[0]
        return group[group[label_col] == selected_label].iloc[0]
    
    # 分组解决冲突
    resolved_df = df.groupby(text_col, group_keys=False).apply(resolve_group)
    return resolved_df.reset_index(drop=True)


resolved_df = resolve_label_conflicts(df)
print("冲突解决结果：")
resolved_df
```
